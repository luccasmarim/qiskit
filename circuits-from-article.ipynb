```python
import pennylane as qml
from random import uniform
from math import pi as pi
import numpy as np
import qutip
import pandas as pd


n_layers = 5
n_wires = 4
n_targets = 10
iteracoes = 10


dev = qml.device('default.qubit', wires=n_wires)
```

```python
def function():


    def normaliza(v):
        v = v / (np.linalg.norm(v))    
        return v

    def cria_alvos(n_targets, n_wires):
        target_states = []
        for i in range(n_targets):
            target_states.append(normaliza(np.random.rand(1, 2**(n_wires))))
        return target_states

    def ket_in_amplt(v):
        aux = []
        for i in range(len(v)):
            v[i] = np.dot(v[i],v[i])
            v[i] = np.linalg.norm(v[i])
            aux.append(float(v[i].real))
        aux = np.array(aux)
        return aux
    

    #Circuitos
    
    @qml.qnode(dev)
    def circuit1(n_layers):
        for i in range(n_layers):
            for i in range(4):
                qml.RX(uniform(0,2*pi),i)
                qml.RZ(uniform(0,2*pi),i)
                
            qml.Barrier()
        return qml.state()

    @qml.qnode(dev)
    def circuit2(n_layers):
        for i in range(n_layers):
            for i in range(4):
                qml.RX(uniform(0,2*pi),i)
                qml.RZ(uniform(0,2*pi),i)

            qml.CNOT(wires=[3,2])
            qml.CNOT(wires=[2,1])
            qml.CNOT(wires=[1,0])

            qml.Barrier() 
        return qml.state()

    @qml.qnode(dev)
    def circuit3(n_layers):
        for i in range(n_layers):
            for i in range(4):
                qml.RX(uniform(0,2*pi),i)
                qml.RZ(uniform(0,2*pi),i)

            qml.CRZ(uniform(0,2*pi),wires=[3,2])
            qml.CRZ(uniform(0,2*pi),wires=[2,1])
            qml.CRZ(uniform(0,2*pi),wires=[1,0])
            
            qml.Barrier()
        return qml.state()

    @qml.qnode(dev)
    def circuit4(n_layers):
        for i in range(n_layers):    
            for i in range(4):
                qml.RX(uniform(0,2*pi),i)
                qml.RZ(uniform(0,2*pi),i)

            qml.CRX(uniform(0,2*pi),wires=[3,2])
            qml.CRX(uniform(0,2*pi),wires=[2,1])
            qml.CRX(uniform(0,2*pi),wires=[1,0])
            
            qml.Barrier()
        return qml.state()
    
    @qml.qnode(dev)
    def circuit5(n_layers):
        for i in range(n_layers):   
            for i in range(4):
                qml.RX(uniform(0,2*pi),i)
                qml.RZ(uniform(0,2*pi),i)

            qml.CRZ(uniform(0,2*pi),wires=[3,2])
            qml.CRZ(uniform(0,2*pi),wires=[3,1])
            qml.CRZ(uniform(0,2*pi),wires=[3,0])

            qml.CRZ(uniform(0,2*pi),wires=[2,3])
            qml.CRZ(uniform(0,2*pi),wires=[2,1])
            qml.CRZ(uniform(0,2*pi),wires=[2,0])

            qml.CRZ(uniform(0,2*pi),wires=[1,3])
            qml.CRZ(uniform(0,2*pi),wires=[1,2])
            qml.CRZ(uniform(0,2*pi),wires=[1,0])

            qml.CRZ(uniform(0,2*pi),wires=[0,3])
            qml.CRZ(uniform(0,2*pi),wires=[0,2])
            qml.CRZ(uniform(0,2*pi),wires=[0,1]) 

            for i in range(4):
                qml.RX(uniform(0,2*pi),i)
                qml.RZ(uniform(0,2*pi),i)

            qml.Barrier()
        return qml.state()    

    @qml.qnode(dev)
    def circuit6(n_layers):
        for i in range(n_layers):     
            for i in range(4):
                qml.RX(uniform(0,2*pi),i)
                qml.RZ(uniform(0,2*pi),i)

            qml.CRX(uniform(0,2*pi),wires=[3,2])
            qml.CRX(uniform(0,2*pi),wires=[3,1])
            qml.CRX(uniform(0,2*pi),wires=[3,0])

            qml.CRX(uniform(0,2*pi),wires=[2,3])
            qml.CRX(uniform(0,2*pi),wires=[2,1])
            qml.CRX(uniform(0,2*pi),wires=[2,0])

            qml.CRX(uniform(0,2*pi),wires=[1,3])
            qml.CRX(uniform(0,2*pi),wires=[1,2])
            qml.CRX(uniform(0,2*pi),wires=[1,0])

            qml.CRX(uniform(0,2*pi),wires=[0,3])
            qml.CRX(uniform(0,2*pi),wires=[0,2])
            qml.CRX(uniform(0,2*pi),wires=[0,1]) 

            for i in range(4):
                qml.RX(uniform(0,2*pi),i)
                qml.RZ(uniform(0,2*pi),i)

            qml.Barrier()

        return qml.state()
    
    @qml.qnode(dev)
    def circuit7(n_layers):
        for i in range(n_layers):    
            for i in range(4):
                qml.RX(uniform(0,2*pi),i)
                qml.RZ(uniform(0,2*pi),i)

            qml.CRZ(uniform(0,2*pi),wires=[1,0])
            qml.CRZ(uniform(0,2*pi),wires=[3,2])

            for i in range(4):
                qml.RX(uniform(0,2*pi),i)
                qml.RZ(uniform(0,2*pi),i)

            qml.CRZ(uniform(0,2*pi),wires=[2,1])

            qml.Barrier()
            
        return qml.state()

    @qml.qnode(dev)
    def circuit8(n_layers):
        for i in range(n_layers):      
            for i in range(4):
                qml.RX(uniform(0,2*pi),i)
                qml.RZ(uniform(0,2*pi),i)

            qml.CRX(uniform(0,2*pi),wires=[1,0])
            qml.CRX(uniform(0,2*pi),wires=[3,2])

            for i in range(4):
                qml.RX(uniform(0,2*pi),i)
                qml.RZ(uniform(0,2*pi),i)

            qml.CRX(uniform(0,2*pi),wires=[2,1])

            qml.Barrier()
            
        return qml.state()

    @qml.qnode(dev)
    def circuit9(n_layers):
        for i in range(n_layers): 
            for i in range(4):
                qml.Hadamard(i)

            qml.SWAP([2,3])
            qml.SWAP([1,2])
            qml.SWAP([0,1])
            
            for i in range(4):
                qml.RX(uniform(0,2*pi),i)

            qml.Barrier()

        return qml.state()

    @qml.qnode(dev)
    def circuit10(n_layers):
        for i in range(n_layers): 
            for i in range(4):
                qml.RY(uniform(0,2*pi),i)

            qml.SWAP([2,3])
            qml.SWAP([1,2])
            qml.SWAP([0,1])
            qml.SWAP([0,3])
            
            for i in range(4):
                qml.RY(uniform(0,2*pi),i)

            qml.Barrier()

        return qml.state()

    @qml.qnode(dev)
    def circuit11(n_layers):
        for i in range(n_layers): 
            for i in range(4):
                qml.RY(uniform(0,2*pi),i)
                qml.RZ(uniform(0,2*pi),i)

            qml.CNOT([3,2])
            qml.CNOT([1,0])

            
            qml.RY(uniform(0,2*pi),1)
            qml.RY(uniform(0,2*pi),2)
            qml.RZ(uniform(0,2*pi),1)
            qml.RZ(uniform(0,2*pi),2)

            qml.CNOT([2,1])

            qml.Barrier()

        return qml.state()
    
    @qml.qnode(dev)
    def circuit12(n_layers):
        for i in range(n_layers):
            for i in range(4):
                qml.RY(uniform(0,2*pi),i)
                qml.RZ(uniform(0,2*pi),i)

            qml.SWAP([3,2])
            qml.SWAP([1,0])

            
            qml.RY(uniform(0,2*pi),1)
            qml.RY(uniform(0,2*pi),2)
            qml.RZ(uniform(0,2*pi),1)
            qml.RZ(uniform(0,2*pi),2)

            qml.SWAP([2,1])

            qml.Barrier()

        return qml.state()

    @qml.qnode(dev)
    def circuit13(n_layers):
        for i in range(n_layers):    
            for i in range(4):
                qml.RY(uniform(0,2*pi),i)

            qml.CRZ(uniform(0,2*pi),[3,0])
            qml.CRZ(uniform(0,2*pi),[2,3]) 
            qml.CRZ(uniform(0,2*pi),[1,2])   
            qml.CRZ(uniform(0,2*pi),[0,1])

            for i in range(4):
                qml.RY(uniform(0,2*pi),i)

            qml.CRZ(uniform(0,2*pi),[3,2])
            qml.CRZ(uniform(0,2*pi),[0,3]) 
            qml.CRZ(uniform(0,2*pi),[1,0])   
            qml.CRZ(uniform(0,2*pi),[2,1])

            qml.Barrier()

        return qml.state()
    
    @qml.qnode(dev)
    def circuit14(n_layers):
        for i in range(n_layers): 
            for i in range(4):
                qml.RY(uniform(0,2*pi),i)

            qml.CRX(uniform(0,2*pi),[3,0])
            qml.CRX(uniform(0,2*pi),[2,3]) 
            qml.CRX(uniform(0,2*pi),[1,2])   
            qml.CRX(uniform(0,2*pi),[0,1])

            for i in range(4):
                qml.RY(uniform(0,2*pi),i)

            qml.CRX(uniform(0,2*pi),[3,2])
            qml.CRX(uniform(0,2*pi),[0,3]) 
            qml.CRX(uniform(0,2*pi),[1,0])  
            qml.CRX(uniform(0,2*pi),[2,1])

            qml.Barrier()

        return qml.state()

    @qml.qnode(dev)
    def circuit15(n_layers):
        for i in range(n_layers): 
            for i in range(4):
                qml.RY(uniform(0,2*pi),i)

            qml.CNOT([3,0])
            qml.CNOT([2,3]) 
            qml.CNOT([1,2])   
            qml.CNOT([0,1])

            for i in range(4):
                qml.RY(uniform(0,2*pi),i)

            qml.CNOT([3,2])
            qml.CNOT([0,3]) 
            qml.CNOT([1,0])   
            qml.CNOT([2,1])

            qml.Barrier()

        return qml.state()

    @qml.qnode(dev)
    def circuit16(n_layers):
        for i in range(n_layers): 
            for i in range(4):
                qml.RX(uniform(0,2*pi),i)

            for i in range(4):
                qml.RZ (uniform(0,2*pi),i)

            qml.CRZ(uniform(0,2*pi),[1,0])
            qml.CRZ(uniform(0,2*pi),[3,2]) 
            qml.CRZ(uniform(0,2*pi),[2,1])   

            qml.Barrier()

        return qml.state()

    @qml.qnode(dev)
    def circuit17(n_layers):
        for i in range(n_layers):   
            for i in range(4):
                qml.RX(uniform(0,2*pi),i)

            for i in range(4):
                qml.RZ(uniform(0,2*pi),i)

            qml.CRX(uniform(0,2*pi),[1,0])
            qml.CRX(uniform(0,2*pi),[3,2]) 
            qml.CRX(uniform(0,2*pi),[2,1])   

            qml.Barrier()

        return qml.state()

    @qml.qnode(dev)
    def circuit18(n_layers):
        for i in range(n_layers): 
            for i in range(4):
                qml.RX(uniform(0,2*pi),i)

            for i in range(4):
                qml.RZ(uniform(0,2*pi),i)

            qml.CRZ(uniform(0,2*pi),[3,0])
            qml.CRZ(uniform(0,2*pi),[2,3]) 
            qml.CRZ(uniform(0,2*pi),[1,2]) 
            qml.CRZ(uniform(0,2*pi),[0,1]) 

            qml.Barrier()

        return qml.state()

    @qml.qnode(dev)
    def circuit19(n_layers):
        for i in range(n_layers): 
            for i in range(4):
                qml.RX(uniform(0,2*pi),i)

            for i in range(4):
                qml.RZ(uniform(0,2*pi),i)

            qml.CRX(uniform(0,2*pi),[3,0])
            qml.CRX(uniform(0,2*pi),[2,3]) 
            qml.CRX(uniform(0,2*pi),[1,2])  
            qml.CRX(uniform(0,2*pi),[0,1]) 

            qml.Barrier()
            
        return qml.state()


    @qml.qnode(dev)
    def circuit20(n_layers):
        qml.RX(uniform(0,2*pi),0)
        
        return qml.state()

    @qml.qnode(dev)
    def circuit21(n_layers): 
        qml.RX(uniform(0,2*pi),0)
        qml.RZ(uniform(0,2*pi),0)
            
        return qml.state()    









    ket = circuit20(n_layers)    #AQUI
    amplt = ket_in_amplt(ket)
    target_states = cria_alvos(n_targets, n_wires)


    #Expressabilidade
    def fidelity(target_states, amplt):
        fidelity_sum = 0
        for target_state in target_states:
            output_probs = amplt
            fidelity = np.vdot(target_state, output_probs) ** 2
            fidelity_sum += fidelity
        average_fidelity = fidelity_sum / len(target_states)
        return average_fidelity
    

    #Grau de Emaranhamento
    def compute_Q_ptrace(ket, N):
        ket = qutip.Qobj(ket, dims=[[2]*(N), [1]*(N)]).unit()
        entanglement_sum = 0
        for k in range(N):
            rho_k_sq = ket.ptrace([k])**2
            entanglement_sum += rho_k_sq.tr()
        Q = 2*(1 - (1/N)*entanglement_sum)
        return Q



    Q = compute_Q_ptrace(ket, n_wires)
    #print("\nGrau de Emaranhamento: {}".format(Q))

    E = fidelity(target_states, amplt)
    #print("Expressabilidade: {}".format(E))

    #print("\n\nEstado gerado: {}".format(ket))
    #print("\n\n\n\n\n")



    k=[Q,E]
    return k





k=[]
for i in range(iteracoes):
    #print("\n=============")
    #print("= Modelo: {} =".format(i+1))
    #print("=============\n")
    k.append(function())


#print("\n\nResultados:\n")
#for i in range(iteracoes):
#    print(" Modelo {} \n Grau de emaranhamento: {} \n Expressabilidade: {} \n".format(i+1,k[i][0],k[i][1]))



def pand(k):
    eman=[]
    exp=[]

    for i in range(iteracoes):
        eman.append(k[i][0])
    for i in range(iteracoes):
        exp.append(k[i][1])

    dic = {"Grau de Emaranhamento": eman, "Expressabilidade":exp}
    df = pd.DataFrame(dic)
    return df

df = pand(k)
df.describe()

```
