# Constructing Quantum Circuits with Quantum Gates

## Importing necessary libraries 

```python
import qiskit
import numpy as np
from qiskit import Aer, QuantumCircuit, execute, ClassicalRegister, QuantumRegister, BasicAer
from qiskit.visualization import plot_histogram
from math import pi, sqrt
```

## NOT Gate

```python
qc_not = QuantumCircuit(1,1,name="qc")

qc_not.x(0)

qc_not.measure(0,0)  

qc_not.draw('mpl')
```

```python
# Use Aer's qasm_simulator
backend = Aer.get_backend('qasm_simulator')

# Execute the circuit on the qasm simulator
job = execute(qc_not, backend, shots=1000)

# Grab results from the job
result = job.result()

# Returns counts
counts = result.get_counts(qc_not)
print("\nTotal counts are:",counts)

# Plot a histogram
plot_histogram(counts)
```

## AND Gate

```python
qc_and = QuantumCircuit(3,1,name="qc")

qc_and.x(0)
qc_and.x(1)

qc_and.ccx(0,1,2)

qc_and.measure(2,0)

qc_and.draw('mpl')
```

```python
# Use Aer's qasm_simulator
backend = Aer.get_backend('qasm_simulator')

# Execute the circuit on the qasm simulator
job = execute(qc_and, backend, shots=1000)

# Grab results from the job
result = job.result()

# Returns counts
counts = result.get_counts(qc_and)
print("\nTotal counts are:",counts)

# Plot a histogram
plot_histogram(counts)
```

## Half Adder Circuit 

```python
# Half Adder Circuit

qc_ha = QuantumCircuit(4,2,name="qc")

# Initialize qubits to |11>
qc_ha.x(0)
qc_ha.x(1)
qc_ha.barrier()

#XOR
qc_ha.cx(1, 2)
qc_ha.cx(0, 2)
qc_ha.barrier()

#AND
qc_ha.ccx(0, 1, 3)
qc_ha.barrier()

#Sum
qc_ha.measure(2, 0)

#Carry out
qc_ha.measure(3, 1)

qc_ha.draw('mpl')
```

```python
# Use Aer's qasm_simulator
backend = Aer.get_backend('qasm_simulator')

# Execute the circuit on the qasm simulator
job = execute(qc_ha, backend, shots=1000)

# Grab results from the job
result = job.result()

# Returns counts
counts = result.get_counts(qc_ha)
print("\nTotal counts are:",counts)

# Plot a histogram
plot_histogram(counts)
```
